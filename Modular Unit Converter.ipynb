```python
Modular Unit Converter:
The Goal: A program that converts temperatures (Celsius to Fahrenheit),
distances (Miles to Kilometers), and weights(Grams to Kilograms).
The Catch: Every single conversion must be its own independent function (e.g., def c_to_f(celsius):). 
    There should be a main menu function that calls the others based on user choice.
Concepts Tested: Defining functions, passing arguments, return statements, variable scope.
```

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
def miles_to_kilometers(miles):
    return miles * 1.609
def grams_to_kilograms(grams):
    return grams / 1000
def main_menu():
    while True:
        print("\nI. Celsius to Fahrenheit")
        print("II. Miles to Kilometers")
        print("III. Grams to Kilograms")
        print("IV. Exit")
        choice = input("Choose I or II or III or IV: ").upper()
        if choice == "I":
            try:
                c = float(input("Enter Celsius: "))
                result = celsius_to_fahrenheit(c)
                print(f"Fahrenheit: {result}")
            except ValueError:
                print("Please enter a valid number!")
        elif choice == "II":
            try:
                m = float(input("Enter Miles: "))
                result = miles_to_kilometers(m)
                print(f"Kilometers: {result}")
            except ValueError:
                print("Please enter a valid number!")
        elif choice == "III":
            try:
                g = float(input("Enter Grams: "))
                result = grams_to_kilograms(g)
                print(f"Kilograms: {result}")
            except ValueError:
                print("Please enter a valid number!")
        elif choice == "IV":
            print("Byebye")
            break
        else:
            print("Invalid choice")
main_menu()
```
