```python
Password Generator:
The Goal: A function that takes an integer (length) and boolean flags (use_numbers, use_symbols) 
and returns a randomized, secure password.
Concepts Tested: Default arguments, keyword arguments, combining lists/strings, returning data.
```

```python
import random
import string
def generate_password(length=10, use_numbers=True, use_symbols=True):
    letters = string.ascii_letters
    numbers = string.digits
    symbols = string.punctuation
    characters = letters
    if use_numbers:
        characters += numbers
    if use_symbols:
        characters += symbols
    password = ""
    for _ in range(length):
        password += random.choice(characters)
    return password    
print(generate_password())  
print(generate_password(20))
print(generate_password(8, use_numbers=False))  
print(generate_password(7, use_symbols=False)) 
print(generate_password(30, use_numbers=False, use_symbols=False)) 
```

```python
import random
import string
def generate_password(length, use_numbers=True, use_symbols=True):
    letters = string.ascii_letters
    numbers = string.digits
    symbols = string.punctuation
    characters = letters
    if use_numbers:
        characters += numbers
    if use_symbols:
        characters += symbols
    password = ""
    for _ in range(length):
        password += random.choice(characters)
    return password
while True:
    try:
        length = int(input("Enter password length: "))
        if length <= 0:
            print("Please enter a positive number.")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter a number.")
while True:
    use_numbers_input = input("Include numbers? (yes/no): ").lower().strip()
    if use_numbers_input in ["yes", "no"]:
        use_numbers = use_numbers_input == "yes"
        break
    else:
        print("Please type yes or no.")
while True:
    use_symbols_input = input("Include symbols? (yes/no): ").lower().strip()
    if use_symbols_input in ["yes", "no"]:
        use_symbols = use_symbols_input == "yes"
        break
    else:
        print("Please type yes or no.")
password = generate_password(length, use_numbers, use_symbols)
print("Generated password:", password)
```
