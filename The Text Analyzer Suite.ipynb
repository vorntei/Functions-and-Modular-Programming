```python
The Text Analyzer Suite
The Goal: Ask the user to input a long sentence or paragraph. 
    Build a program that outputs statistics about their text.
The Catch: You must use separate functions for every analysis: 
count_words(text), count_vowels(text), and find_longest_word(text). 
    Each function must take the text as an argument and return the result to be printed by the main program.
Concepts Tested: String manipulation, passing strings as arguments, 
returning specific values (integers and strings) from functions, and keeping logic strictly separated.
```

```python
def count_words(text):
    words = text.split()
    return len(words)
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
     return count
def find_longest_word(text):
    words = text.split()
    longest = "" #ah nis vea do pe rk see longest word
    for word in words: #rk mer mdg 1 peak
        if len(word) > len(longest): #len(word)=aksor in peak tmey, len(longest)=aksor in the longest word ah jas
            longest = word #ber peak ng vea veng jeang logest ah jas replace dak ah tmey nis vnh
    return longest
def main():
    text = input("Enter a sentence or paragraph: ")
    words_count = count_words(text)
    vowels_count = count_vowels(text)
    longest_word = find_longest_word(text)
    print("Number of words:", words_count)
    print("Number of vowels:", vowels_count)
    print("Longest word:", longest_word)
main()
```
