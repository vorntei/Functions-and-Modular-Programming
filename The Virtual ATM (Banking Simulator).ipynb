```python
The Virtual ATM (Banking Simulator)
The Goal: Simulate a bank account where a user can deposit, withdraw, or check their balance.
The Catch: Create specific functions like deposit(balance, amount) and withdraw(balance, amount). 
    These functions must return the newly calculated balance to overwrite the old balance in the main program loop. 
    The withdraw function must also include logic to prevent overdrafting 
        (returning an error message and the unmodified balance if they ask for too much).
Concepts Tested: Passing variables as arguments,
    updating state using return statements, handling edge cases within functions,
    and understanding variable scope (realizing that a variable modified inside a function doesn't change outside unless returned).
```

```python
def deposit(balance, amount):
    balance += amount
    return balance
def withdraw(balance, amount):
    if amount > balance:
        print("luy ot krup te")
        return balance
    else:
        balance -= amount
        return balance
def main():
    balance = 1000
    while True:
        print("Virtual ATM")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")
        #deposit=dak luy, withdraw=dok luy
        choice = input("Choose 1 2 3 4: ")
        if choice == "1":
            amount = float(input("Enter deposit amount: "))
            if amount > 0:
                balance = deposit(balance, amount)
                print("Deposit successful.")
            else:
                print("amount ot trv")
        elif choice == "2":
            amount = float(input("Enter withdrawal amount: "))
            if amount > 0:
                balance = withdraw(balance, amount)
            else:
                print("ot luy dol pin ng te")
        elif choice == "3":
            print("Current balance:", balance)
        elif choice == "4":
            print("byebye")
            break
        else:
            print("ah ng ot mean in choice te")
if __name__ == "__main__":
    main()
```
